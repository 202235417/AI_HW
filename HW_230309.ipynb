#1. 필요 도구들 임포트 / 변수 정의


```python
from sklearn.datasets import load_breast_cancer

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.neighbors import KNeighborsClassifier

breast_cancer_data = load_breast_cancer()
df_features = breast_cancer_data.feature_names
df_data = pd.DataFrame(breast_cancer_data.data, columns = df_features)
df_labels = pd.DataFrame(breast_cancer_data.target)  

print(breast_cancer_data.target_names)
X_train, X_test, y_train, y_test = train_test_split(df_data, df_labels, train_size=0.8, random_state=1000)

print("X_train.shape :", X_train.shape)

print("y_train.shape :", y_train.shape)

print("X_test.shape:", X_test.shape)

print("y_test.shape:", y_train.shape)

mymodel = KNeighborsClassifier(n_neighbors = 3)
mymodel.fit(X_train, y_train)


# 예측
prediction = mymodel.predict(X_test)
mymodel.score(X_test, y_test)
```

#3. 결측치 확인(Null값 확인)

```python
#결측치 확인 - isnull()
#isnull() : 결측치가 있으면 True, 아니면 False를 나타내는 함수.

#isnull()
print('==================================isnull()로 결측치 확인==================================')
df_data.isnull()
```

```python
#결측치 확인 - isnull().sum()
#isnull().sum() : 결측치의 개수를 알려주는 함수.

print('==================================isnull().sum()으로 결측치 확인==================================')
df_data.isnull().sum() 
```

> isnull()로는 False가, isnull().sum()로는 모든 데이터가 0으로 출력되므로 이 데이터에는 결측치가 없다.
