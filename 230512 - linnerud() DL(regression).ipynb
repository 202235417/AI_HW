```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats  
from sklearn.datasets import load_linnerud
```

```python
dataset = load_linnerud()
```

```python
dataset
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.target = pd.DataFrame(dataset.target, columns = dataset.target_names)
```

```python
df
```

```python
df.target
```

```python
data = pd.concat([df, df.target], axis=1)
```

```python
data.drop(columns=['Pulse', 'Waist'], inplace=True)
```

```python
data
```

```python
data['Weight'] = dataset.target
```

```python
data
```

```python
X=data.drop('Weight',axis=1)
X.head()
```

```python
y = data['Weight']
y.head()
```

```python
Y = pd.get_dummies(data['Weight']).values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
# 평균
mean = np.mean(X_train, axis = 0)

# 표준편차
std = np.std(X_train, axis = 0) # 표준편차값

# 표준화값
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(1024, activation = 'relu', input_shape = (3,)))
model.add(Dense(512, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1)) 
model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mae'])

model.summary()
```

```python
N_EPOCHS = 100
history = model.fit(X_train, y_train, epochs = N_EPOCHS, validation_data=(X_test, y_test))
```

```python
# loss 체크
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# loss 체크
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['mae'], label='mse')
plt.plot(range(N_EPOCHS), history.history['val_mae'], label='test mse')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
